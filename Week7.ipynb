```python
# Iris Dataset Analysis in Python
# Import necessary libraries - step 1
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import urllib.request
```

```python
# Download the dataset - step 2
url = "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
filename = "iris.csv"
```

```python
try:
    urllib.request.urlretrieve(url, filename)
    print("Dataset downloaded successfully.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# Defining the column names - step 3
column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
```

```python
# Loadng the dataset - step 4
try:
    iris_df = pd.read_csv('iris.csv', header=None, names=column_names)
    print("Dataset loaded successfully.")
except FileNotFoundError:
    print("Error: File not found. Make sure 'iris.csv' is in your working directory.")
```

```python
# Display first 5 few rows - step 5
print("\nFirst five rows of the dataset:")
print(iris_df.head())
```

```python
# Explore structure - step 6
print("\nData types:")
```

```python
print(iris_df.dtypes)
```

```python
# missing values
print("\nMissing values:")
print(iris_df.isnull().sum())
```

```python
# Basic statistics for the dataset - step 7
print("\nBasic statistical summary:")
print(iris_df.describe())
```

```python
# Grouping by species - step 8
print("\nAverage measurements per species:")
print(iris_df.groupby('species').mean())
```

```python
# Data Visualization - Step 9
sns.set(style="whitegrid")
```

```python
# Line chart (simulated using index as time)
plt.figure(figsize=(10, 6))
iris_df.groupby('species').mean().plot(marker='o')
plt.title('Average Measurements per Species')
plt.ylabel('cm')
plt.xlabel('Species')
plt.legend(title='Features')
plt.grid(True)
plt.show()
```

```python
# Bar chart
plt.figure(figsize=(8, 6))
iris_df.groupby('species')['petal_length'].mean().plot(kind='bar', color='skyblue')
plt.title('Average Petal Length per Species')
plt.ylabel('Petal Length (cm)')
plt.xlabel('Species')
plt.show()
```

```python
# Histogram
plt.figure(figsize=(8, 6))
sns.histplot(iris_df['sepal_length'], kde=True, bins=20, color='green')
plt.title('Distribution of Sepal Length')
plt.xlabel('Sepal Length (cm)')
plt.show()
```

```python
# Scatter plot
plt.figure(figsize=(8, 6))
sns.scatterplot(data=iris_df, x='sepal_length', y='petal_length', hue='species', palette='deep')
plt.title('Sepal Length vs Petal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.show()
```
